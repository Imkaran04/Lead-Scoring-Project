# import necessary libraries 

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
from sklearn.feature_selection import SelectKBest, chi2
```

```python
# loading the dataset 
df = pd.read_csv("Data_Science_Internship.csv")
df.head(10)
```

# Data Cleaning

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

# Feature selection

```python
cat_cols = ['Agent_id', 'status', 'lost_reason', 'budget', 'lease', 'movein', 'source', 'source_city', 'source_country', 'utm_source', 'utm_medium', 'des_city', 'des_country', 'room_type']
df_cat = df[cat_cols]
```

```python
df.info()
```

# Data Preprocessing

```python

df_cat_encoded = pd.get_dummies(df_cat, drop_first=True)
```

# Train Test split

```python
df_cat_encoded.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_cat_encoded.drop('Agent_id_06ae80c3e00a1457fc70192d357766c9de0a9787fe3dedc386ac0191e9c151c3', axis=1), df_cat_encoded['Agent_id_06ae80c3e00a1457fc70192d357766c9de0a9787fe3dedc386ac0191e9c151c3'], test_size=0.2, random_state=42)
```

# Training the model

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
```

# testing the model

```python
y_pred = rf.predict(X_test)
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```

# Evaluting the model

```python
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```

# Saving the model

```python
import pickle

# save the model to disk
filename = 'random_forest_model.pkl'
pickle.dump(RandomForestClassifier, open(filename, 'wb'))

```
